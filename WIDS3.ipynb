```python
import numpy as np
import pandas as pd

Users = pd.read_csv(r"C:\Users\Asus\OneDrive\Documents\OneNote Notebooks\wids3.psv", sep="|")
Users.head(10)
```

```python
Users.info()
```

```python
MM=Users.groupby('occupation' , as_index=True)['age'].mean()
MM.head(10)
```

```python
MR=Users.groupby('occupation' , as_index=True)['gender']\
                  .apply(lambda x: (x == 'M').mean())
MR.sort_values(ascending=False)
```

```python
MR=Users.groupby('occupation' , as_index=True)['gender']\
                  .agg( male_ratio=lambda x: (x == 'M').mean())
MR.head()
```

```python
MR=Users.groupby('occupation' , as_index=False)['age'].agg(['max', 'min'])
MR
                  
```

```python
MA=Users.groupby(['occupation' ,'gender' ], as_index=False)['age'].mean()
MA.head(10)
```

```python
gender_ratio = Users.groupby('occupation', as_index=False)['gender'] \
    .agg(
        male_ratio=lambda x: (x.str.lower().str.strip() == 'm').mean(),
        female_ratio=lambda x: (x.str.lower().str.strip() == 'f').mean()
    )

gender_ratio.head()

```
