```python
# Cell 6: Transpose and flatten an array
y = np.array([[1,2,3],[4,5,6]])

y_transpose = y.reshape((3,2)).T  # Using reshape method
y_flattened = y.reshape(-1)

print(y_transpose)
print(y_flattened)

# Direct numpy methods
y_transpose = y.T
y_flattened = y.flatten()
print(y_transpose)
print(y_flattened)
```

```python
# Cell 9: Create a linspace from 0.1 to 0.9 with 10 points
start = 0.1
end = 0.9
arr_linspace = np.linspace(start, end, num=10, endpoint=True, axis=0)
print(arr_linspace)
```

```python
# Cell 10: Slicing examples
arr1 = np.array([1,2,3,4,5])
print(arr1[0:5])
print(arr1[0:4])
print(arr1[0:4:1])
print(arr1[0:4:2])
print(arr1[::-1])

arr = np.arange(40).reshape(4,-1)
print(arr[1:4,2:5])
```

```python
# Cell 11: Copies vs references
arr = np.random.randint(0,10,size=6)
print("Original arr:", arr)

new_arr = arr
arr[0] = 100
print("Modified arr:", arr)
print("Reference new_arr:", new_arr)

# Create a true copy
arr = np.random.randint(0,10,size=6)
new_arr = arr.copy()
arr[0] = -39
print("Original arr after copy modification:", arr)
print("Copied new_arr:", new_arr)
```

```python
# Cell 12: Random number generation
arr = np.random.rand(3,4)
print(arr)

# Fixing seed to get same results every time
np.random.seed(31)
arr = np.random.rand(3,4)
print(arr)

# Random integers from [5,12) with shape (2,4)
arr_randint = np.random.randint(5, 13, size=(2,4))
print(arr_randint)
```

```python
# Cell 13: Aggregation functions
arr = np.arange(10).reshape(2,5)
print(arr.sum())
print(arr.mean())
print(arr.std())
print(arr.min())

arr = np.array([[1,2,9],
                [3,4,5],
                [-3,8,7]])
print(arr)
print(arr.max(axis=0))
print(arr.max(axis=1))
print(arr.max(), arr.min(), arr.mean())
```

```python
# Cell 14: Broadcasting and arithmetic operations
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
print("arr1:", arr1)
print("Scalar multiplication:", arr1*2)
print("Element-wise multiplication:", arr1*arr2)

arr_square = arr1**2
print("Square of each element:", arr_square)

arr = np.arange(12).reshape(3,4)
arr1 = np.array([1,2,3,4])
arr2 = np.array([7,2,5,8])
print("Addition:", arr+1)
print("Subtraction:", arr-1)
print("Multiply by 2:", arr*2)
print("Division by 2:", arr/2)
print("Integer division:", arr//2)
print("Power 3:", arr**3)
print("Element-wise multiplication arr*arr1:", arr*arr1)
print("Element-wise multiplication arr*arr2:", arr*arr2)
```
